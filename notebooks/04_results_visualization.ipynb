# Results Visualization

Visualize federated learning results and blockchain.

```python
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load blockchain ledger
with open('../blockchain_ledger.json', 'r') as f:
    blockchain = json.load(f)

print(f"Total blocks: {len(blockchain)}")
```

```python
# Extract FL rounds
fl_rounds = [b for b in blockchain if b.get('data', {}).get('type') == 'fl_round']

# Plot metrics over rounds
rounds = [r['data']['round'] for r in fl_rounds]
accuracies = [r['data']['metrics'].get('accuracy', 0) for r in fl_rounds]

plt.figure(figsize=(10, 6))
plt.plot(rounds, accuracies, marker='o')
plt.xlabel('FL Round')
plt.ylabel('Global Accuracy')
plt.title('Federated Learning Progress')
plt.grid(True)
plt.show()
```
